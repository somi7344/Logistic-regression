```python
devicetype = OneHotEncoder(sparse_output = False)
device_encoded = devicetype.fit_transform(data[['Device_Type']])
device_encoded 
```

```python
device_encoded_df = pd.DataFrame(device_encoded, columns=devicetype.get_feature_names_out(["Device_Type"]))
device_encoded_df
```

```python
ftt = pd.concat([data,device_encoded_df,occupation_encoded_df,gender_encoded],axis=1).drop(columns=['Device_Type','Gender','Occupation'])
```

```python
ftt.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns to scale
num_cols = ['Age', 'Daily_Phone_Hours', 'Social_Media_Hours',
            'Work_Productivity_Score', 'Sleep_Hours',
            'App_Usage_Count', 'Caffeine_Intake_Cups', 'Weekend_Screen_Time_Hours']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(ftt[num_cols])  # only these columns
```

```python
# Select all the binary columns
binary_cols = ['Device_Type_Android', 'Device_Type_iOS',
               'Occupation_Freelancer', 'Occupation_Professional', 'Occupation_Student',
               'Female', 'Male', 'Other']

# Combine scaled numerics + binary columns
X_final = pd.concat([pd.DataFrame(X_scaled, columns=num_cols), ftt[binary_cols].reset_index(drop=True)], axis=1)

# Target
Y = ftt['Stress_Level']
```

```python
X_final.columns
```

```python
ftt["Stress_Level"] = ftt["Stress_Level"].apply(lambda x: 1 if x >= 3 else 0)
```

```python
from sklearn.model_selection import train_test_split

X = X_final
Y = ftt['Stress_Level']

X_train, X_test, Y_train, Y_test = train_test_split(
    X_final,Y, test_size = 0.2, random_state = 42)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_compare = pd.DataFrame(y_pred,Y_test)
y_compare
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
```
